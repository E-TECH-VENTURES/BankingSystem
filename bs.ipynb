```python
class BankingSystem:
    AccountPassDict = {}
    c_last = ""
    s_last = ""
    def __init__(self):
        
        with open("_confidential.txt","r") as updateDict:
            if self.readLastLine(updateDict) == '0':
                self.AccountPassDict = {}
            else:
                updateDict.seek(0)
                lines = updateDict.read()
                n = False
                c = 0
                _acc = ''
                _pas = ''
                for data in lines:
                    if c == 0:
                        pass
                    else:
                        self.AccountPassDict[_acc] = _pas
                    if n == False:
                        _acc = data
                        n = True
                    if n == True:
                        _pas = data
                        n = False
        with open("c_last.txt", "r") as cLastRead:
            if self.readLastLine(cLastRead) == "0":
                with open("c_last.txt","w") as cLastWrite:
                    cLastWrite.write("24001000")
            else:
                 self.c_last = str(self.readLastLine(cLastRead))      # Default values already set
                                                               # will update these numbers as per last saved info
                        
        with open("s_last.txt", "r") as sLastRead:
            if self.readLastLine(sLastRead) == "0":
                with open("s_last.txt","w") as sLastWrite:
                    sLastWrite.write("26001000")
            else:
                self.s_last = str(self.readLastLine(sLastRead))
        #c_last = "24001000"            #Current Account Numbers range
        #s_last = "26001000"            #Saving Account Numbers range
        
    def nowTime(self):
        return datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
     
    def readLastLine(self, file):
        lines = file.read().splitlines()
        try:
            last_line = lines[-1]
        except IndexError:
            last_line = "0"
        return last_line
    def authenticate(self,msg):
        # ask admin if he authenticate the msg
        ask = input("\nFor ADMIN\n\tDo you authenticate this "+msg+": ")
        if ask.lower() == "yes":
            return True
        return False
```

```python
class Manage(Accounts,BankingSystem): #AD removed 
    def applyForCard(self):
        account = input("Enter account number: ")
        print("\n400 for your card has been deducted from your account\n\n")
        
        with open(account+".txt","r") as applyForCardRead:
            bal = int(self.readLastLine(applyForCardRead))
        if bal >= 400:
            with open(account+".txt","a") as applyCard:
                applyCard.write("\n"+"Credit Card"+"\n"+str(self.nowTime())+"\n"+str(bal-400))
        else:
            print("\nNot enough balance to apply for card!\n")
    def applyForCheque(self,account = None,noOfpages = None):
        account = input("Enter account number: ")
        noOfpages = int(input("Enter number of Pages of Cheque: "))
        print("\nRs.10/- per page for your cheque has been deducted from your account\n\n")
        
        
        with open(account+".txt","r") as applyChequeRead:
            bal = int(self.readLastLine(applyChequeRead))
        if bal >= 10*int(noOfpages):
            with open(account+".txt","a") as applyCheque:
                applyCheque.write("\n"+"Cheque of "+str(noOfpages)+"\n"+str(self.nowTime())+"\n"+str(bal-10*noOfpages))
        else:
            print("\nNot enough balance to apply for Cheque!\n")
        
        
       
    def BalanceSheet(self,account = None):
        account = input("Enter account number: ")
        with open(account+".txt","r") as BalSheet:
            lines = BalSheet.read().splitlines()
            for a in lines:
                print(a)
    def payOrder(self,account=None,amount=None):
        account = input("Enter account number: ")
        amount = int(input("Enter amount of payorder: "))
        if self.authenticate("Payorder"):
            with open(account+".txt","r") as payOrderRead:
                bal = int(self.readLastLine(payOrderRead))
            with open(account+".txt","a") as payOrder:
                if bal >= int(amount):
                    payOrder.write("\nPay Order generated"+"\n"+str(self.nowTime())+"\n"+str(amount)+"\n"+str(int(bal) - int(amount)))
                else:
                    print("\nNot enough balance!\n")
            print("\nPay Order for Rs."+str(amount)+" is generated!\n")
        else:
            print("\nAuthorization Failed!\n")
```

```python
class LoanFacility(Manage,Accounts,BankingSystem):  #AD removed
    def __init__(self):
        self.loanGranted = 0
    def applyForLoan(self,account = None,amount = None):
        account = input("Enter account number: ")
        amount = int(input("Enter amount of you want: "))
        with open(str(account)+".txt","r") as loanApplyRead:
            bal = int(self.readLastLine(loanApplyRead))
        with open(str(account)+".txt","a") as loanApply:
            lastLine = bal
            amount_against_loan = lastLine
            a = int(input("Enter worth of items you have submitted to bank: "))
            amount_against_loan += a
            if amount_against_loan >= amount:
                print("\nLoan in granting condition!\n")
                if self.authenticate("Loan of "+ str(amount)):
                    loanApply.write("\nLoan Granted!\n"+str(amount)+"\nWhen items worth "+str(amount_against_loan)+" were allocated to bank"+"\n"+str(self.nowTime())+"\nBalance After Loan Granted:\n"+str(amount)+"\n"+str(int(lastLine)+int(amount)))
                    print("\nLoan of "+amount+" is Granted to you!\n")
                else:
                    print("Sorry Sir, Your request for loan is rejected!")
            else:
                print("Dear Sir, you need to re-apply for loan with more worthy items than you just deposited at bank. See you soon!")
        return
    def loanInstallmentsDetails(self,account = None,amountOfLoan = None):
        account = input("Enter account number: ")
        amountOfLoan = int(input("Enter amount of loan you were given : "))
        time = float(input("Enter duration in months you will pay your loan: "))
        split = int(input("Enter duration for your 2 consecutive installments in month(s): "))
        amountPerAttemp = int(amountOfLoan) / (time )
        print("Dear Sir, as per your request, now you have to pay "+str(amountPerAttemp)+" every "+ str(split) +" months\n")
        return amountPerAttemp
    def LoanCollection(self,account = None,amount = None):
        account = input("Enter account number: ")
        choose = input("Do you want pay by cash or want to deduct through your remaing balance: (cash or remaining)")
        if choose.lower() == "cash" or choose[0].lower() == 'c':
            amount = int(input("Enter Amount you have to pay for loan: "))
            with open(str(account)+".txt","r") as loanCollectRead:
                lines = loanCollectRead.read().splitlines()
                loanremaining = lines[-2]
            with open(str(account)+".txt","a") as loanCollect:
                loanCollect.write("\nLoan Installment Paid\n"+str(amount)+"\n"+str(self.nowTime())+"\n"+str(int(loanremaining)-int(amount))+"\n"+str(lines[-1]))
                print("Your amount for "+ str(amount)+ " is accepted against "+ str(loanremaining) +" paid by cash "+ "\nNow you have to pay ",str(int(loanremaining)-amount))
        elif choose.lower() == "remaining" or choose[0].lower() == 'r':
            amount = int(input("Enter Amount you have to pay for loan: "))
            with open(str(account)+".txt","r") as loanCollectRead:
                lines = loanCollectRead.read().splitlines()
                loanremaining = lines[-2]
                # bal = self.readLastLine(loanCollectRead) 
                bal = lines[-1]
                if int(bal) >= int(amount):
                    with open(str(account)+".txt","a") as loanCollect:
                        loanCollect.write("\nLoan Installment Paid\n"+str(amount)+"\n"+str(self.nowTime())+"\n"+str(int(loanremaining)-int(amount))+"\n"+str(int(bal) - int(amount)))
                        print("Your amount for "+ str(amount)+ " is accepted against "+ str(loanremaining) + "\nNow you have to pay "+str(int(loanremaining)-amount)+"\nYour remaing balance in your account is: "+str(int(bal) - int(amount)))
                else:
                    print("\nYou don't have enough balance!\n")
        return
    
```
