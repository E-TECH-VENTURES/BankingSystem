```python
class BankingSystem:
    AccountPassDict = {}
    c_last = ""
    s_last = ""
    def __init__(self):
        with open("c_last.txt", "r") as cLastRead:
            if self.readLastLine(cLastRead) == "":
                with open("c_last.txt","w") as cLastWrite:
                    cLastWrite.write("\n24001000")
        with open("s_last.txt", "r") as sLastRead:
            if self.readLastLine(sLastRead) == "":
                with open("s_last.txt","w") as sLastWrite:
                    sLastWrite.write("\n26001000")
        with open("c_last.txt","r") as cl:           # Default values already set
            self.c_last = str(self.readLastLine(cl))                   # will update these numbers as per last saved info
        with open("s_last.txt","r") as sl:
            self.s_last = str(self.readLastLine(sl))
        #c_last = "24001000"            #Current Account Numbers range
        #s_last = "26001000"            #Saving Account Numbers range
        
    def nowTime(self):
        return datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
     
    def readLastLine(self, file):
        lines = file.read().splitlines()
        try:
            last_line = lines[-1]
        except IndexError:
            last_line = "0"
        print(last_line)
        return last_line
    def authenticate(self,msg):
        # ask admin if he authenticate the msg
        ask = input("\nFor ADMIN\n\tDo you authenticate this "+msg+": ")
        if ask.lower() == "yes":
            return True
        return False
```

```python
class LoanFacility(Manage,Accounts,BankingSystem):  #AD removed
    def __init__(self):
        self.loanGranted = 0
    def applyForLoan(self,account = None,amount = None):
        account = input("Enter account number: ")
        amount = int(input("Enter amount of you want: "))
        with open(str(account)+".txt","w+") as loanApply:
            lastLine = int(self.readLastLine(loanApply))
            amount_for_loan = lastLine
            a = int(input("Enter worth of items you have submitted to bank: "))
            amount_for_loan += a
            if amount_for_loan >= amount:
                print("\nLoan in granting condition!\n")
                if self.authenticate("Loan of "+ str(amount)):
                    loanApply.write("\nLoan Granted!\n"+str(amount)+"\nWhen items worth "+str(amount_for_loan)+" were allocated to bank\n"+"\n"+str(self.nowTime())+"\nBalance After Loan Granted:\n"+str(amount)+"\n"+str(int(lastLine)+int(amount)))
                else:
                    print("Sorry Sir, Your request for loan is rejected!")
            else:
                print("Dear Sir, you need to re-apply for loan with more worthy items than you just deposited at bank. See you soon!")
        return
    def loanInstallmentsDetails(self,account = None,amountOfLoan = None):
        account = input("Enter account number: ")
        amountOfLoan = int(input("Enter amount of loan you were given : "))
        time = float(input("Enter duration in months you will pay your loan: "))
        split = int(input("Enter duration for your 2 consecutive installments in month(s): "))
        amountPerAttemp = int(amountOfLoan) / (time )
        print("Dear Sir, as per your request, now you have to pay "+str(amountPerAttemp)+" every"+ str(split) +"months\n")
        return amountPerAttemp
    def LoanCollection(self,account = None,amount = None):
        account = input("Enter account number: ")
        amount = int(input("Enter Amount you have paid for loan: "))
        with open(str(account)+".txt","w+") as loanCollect:
            lines = loanCollect.read().splitlines()
            loanremaining = lines[-2]
            loanCollect.write("\nLoan Installment Paid\n"+str(amount)+"\n"+self.nowTime()+"\n"+str(loanremaining-int(amount))+"\n"+lines[-1])
            print("Your amount for "+ str(amount)+ " is accepted against "+ str(loanremaining) + "\nNow you have to pay ",str(int(loanremaining)-amount))
        return
    
```
